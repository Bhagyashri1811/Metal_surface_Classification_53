```python
import plotly.express as px
import os
import skimage.io
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm, datasets
```

```python
x_train = list()
y_train = list()

```

```python
import glob
import cv2


for directory_path in glob.glob("/content/drive/MyDrive/Colab Notebooks/images/train/*"):
  label = directory_path.split("\\")[-1]
  print(label)
  for img_path in glob.glob(os.path.join(directory_path,"*.jpg")):
    print(img_path)
    img= cv2.imread(img_path,cv2.IMREAD_COLOR)
    img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
    img = img.reshape(-1)
    x_train.append(img)#append all the image in imge
    y_train.append(label)#append all the class in classes
```

```python
x_test = list()
y_test = list()

for directory_path in glob.glob("/content/drive/MyDrive/Colab Notebooks/images/test/*"):
  label = directory_path.split("\\")[-1]
  print(label)
  for img_path in glob.glob(os.path.join(directory_path,"*.jpg")):
    print(img_path)
    img= cv2.imread(img_path,cv2.IMREAD_COLOR)
    img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
    img = img.reshape(-1)
    x_test.append(img)#append all the image in imge
    y_test.append(label)#append all the class in classes
```

```python
import numpy as np
x_train=np.array(x_train)
y_train=np.array(y_train)
x_test=np.array(x_test)
y_test=np.array(y_test)
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(y_train)
y_train = le.transform(y_train)
le.fit(y_test)
y_test = le.transform(y_test)
```

```python
from sklearn import svm
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
model = svm.SVC()
model.fit(x_train, y_train)
 
# print prediction results
predictions = model.predict(x_test)
print(classification_report(y_test, predictions))
```

```python
from sklearn.model_selection import GridSearchCV

# defining parameter range
param_grid = {'C': [0.1, 1, 10, 100, 1000],
			'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
			'kernel': ['rbf','linear','poly','sigmoid']}

grid = GridSearchCV(model, param_grid, refit = True, verbose = 3)

# fitting the model for grid search
grid.fit(x_train, y_train)
```

```python
# print best parameter after tuning
print(grid.best_params_)

# print how our model looks after hyper-parameter tuning
print(grid.best_estimator_)
```

```python
grid_predictions = grid.predict(x_test)

# print classification report
print(classification_report(y_test, grid_predictions))
```

```python
linear = svm.SVC(kernel='linear', C=1).fit(x_train, y_train)
y_pred=linear.predict(x_test)
#print("Linear Classifier -->",accuracy_score(y_test,y_pred))
#print(confusion_matrix(y_test,y_pred))
#print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(accuracy_score(y_test,y_pred))
```

```python
rbf = svm.SVC(kernel='rbf', gamma=0.5, C=1).fit(x_train, y_train)
y_pred=rbf.predict(x_test)
print("RBF -->",accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
poly = svm.SVC(kernel='poly', degree=5, C=1).fit(x_train, y_train)
y_pred=poly.predict(x_test)
print("Polynomial Classifier-->",accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
sig = svm.SVC(kernel='sigmoid', C=1).fit(x_train, y_train)
y_pred=sig.predict(x_test)
print("Sigmoidal Classifier-->",accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
