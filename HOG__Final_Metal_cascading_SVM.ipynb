```python
import plotly.express as px
import os
import skimage.io
import numpy as np
import matplotlib.pyplot as plt
```

```python
#importing required libraries
from skimage.io import imread, imshow
from skimage.transform import resize
from skimage.feature import hog
from skimage import exposure
%matplotlib inline
```

```python
x_train = list()
y_train = list()

```

```python
import glob
import cv2


for directory_path in glob.glob("/content/drive/MyDrive/Colab Notebooks/images/train/*"):
  label = directory_path.split("\\")[-1]
  print(label)
  for img_path in glob.glob(os.path.join(directory_path,"*.jpg")):
    print(img_path)
    img= cv2.imread(img_path,cv2.IMREAD_COLOR)
    img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
    #img = img.reshape(-1)
    fd, hog_image = hog(img, orientations=9, pixels_per_cell=(8, 8), 
                    cells_per_block=(2, 2), visualize=True, multichannel=True)
    hog_image_rescaled = exposure.rescale_intensity(hog_image, in_range=(0, 10))
    x_train.append(hog_image_rescaled)#append all the image in imge
    y_train.append(label)#append all the class in classes
```

```python
x_test = list()
y_test = list()

for directory_path in glob.glob("/content/drive/MyDrive/Colab Notebooks/images/test/*"):
  label = directory_path.split("\\")[-1]
  print(label)
  for img_path in glob.glob(os.path.join(directory_path,"*.jpg")):
    print(img_path)
    img= cv2.imread(img_path,cv2.IMREAD_COLOR)
    img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
    #img = img.reshape(-1)
    fd, hog_image = hog(img, orientations=9, pixels_per_cell=(8, 8), 
                    cells_per_block=(2, 2), visualize=True, multichannel=True)
    hog_image_rescaled = exposure.rescale_intensity(hog_image, in_range=(0, 10))
    x_test.append(hog_image_rescaled)#append all the image in imge
    y_test.append(label)#append all the class in classes
```

```python
import numpy as np
x_train=np.array(x_train)
y_train=np.array(y_train)
x_test=np.array(x_test)
y_test=np.array(y_test)
```

```python
nsamples, nx, ny = x_train.shape
x_train = x_train.reshape((nsamples,nx*ny))
nsamples, nx, ny = x_test.shape
x_test = x_test.reshape((nsamples,nx*ny))
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(y_train)
y_train = le.transform(y_train)
le.fit(y_test)
y_test = le.transform(y_test)
```

```python
from sklearn import svm
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
model = svm.SVC()
model.fit(x_train, y_train)
 
# print prediction results
predictions = model.predict(x_test)
print(classification_report(y_test, predictions))
```

```python
poly = svm.SVC(kernel="poly")
poly.fit(x_train, y_train)
```

```python
from sklearn.model_selection import GridSearchCV

# defining parameter range
#param_grid = {'C': [1, 10, 100, 1000]}
param_grid = {'kernel': ['rbf','linear','poly','sigmoid']}
param_grid_1 = {'degree': [1, 2, 3,4,5,6,7,8,9]}


grid = GridSearchCV(model, param_grid, refit = True, verbose = 3)
grid_poly = GridSearchCV(poly, param_grid_1, refit = True, verbose = 3)
# fitting the model for grid search
grid.fit(x_train, y_train)
```

```python
grid_poly.fit(x_train, y_train)
```

```python
# print best parameter after tuning
print(grid.best_params_)

# print how our model looks after hyper-parameter tuning
print(grid.best_estimator_)
```

```python
# print best parameter after tuning
print(grid_poly.best_params_)

# print how our model looks after hyper-parameter tuning
print(grid_poly.best_estimator_)
```

```python
poly = svm.SVC(kernel='poly', degree=2, C=1).fit(x_train, y_train)
y_pred=poly.predict(x_test)
print("Polynomial Classifier-->",accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
rbf = svm.SVC(kernel='rbf', C=100).fit(x_train, y_train)
y_pred=rbf.predict(x_test)
print("RBF Classifier-->",accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
